## 6. Realizar gráficas
Para poder tener mejor visualización de los datos, vamos a realizar algunas gráficas. 

#### 6.1 Videojuegos realizados por año. 

##### Grafica de Barras. 
Ya que nos interesa primero tener en cuenta cuanta de nuestra información esta dentro del periodo de tiempo que buscamos.  
> Realizados del 2015 a la fecha.  

Hacemos una grafica de barram de cuantos videojuegos salieron al mercado por año. 

```python
dataset_games_filter = dataset_games[dataset_games['Year_of_Release'] >= 1980] #Filtramos los años, ya que hay había datos sin valor en esta columna., 

year_counts = dataset_games_filter.groupby("Year_of_Release").size()
plt.bar(year_counts.index, year_counts.values)
plt.title("Number of Video Game Releases per Year")
plt.xlabel("Year")
plt.ylabel("Number of Releases")
plt.show()

```

Podemos ver que la mayoría de la información con la que contamos se encuentra entre el año 2000 y 2015.

##### Diagrama de caja y bigotes. 

Para analizar y visualizarde una mejor manera nuetra variable año, hacemos un diagrama de caja bigotes. 
El diagrama nos ayudara a dividir nuetra información en cuartiles. 

```python
import seaborn as sns

sns.boxplot(x=dataset_games_filter["Year_of_Release"])
```

El diagrama nos ha marcado varios outliers, lo cuál puede deberse a que en los primeros años, la realizaciónd e videojuegos era muy pequeña. 

El Rango intercuatil nos dice que la mayoría de la información (50% de lso datos) se concentran entre el 2003 al 2010 aproximadamente, por lo cuál la información queda muy reducida de nuectro dataset. 
 

```python
filter_year = dataset_games[dataset_games['Year_of_Release'] >= 2015] #Filtramos los años, ya que hay había datos sin valor en esta columna., 

year_counts = filter_year.groupby("Year_of_Release").size()
plt.bar(year_counts.index, year_counts.values)
plt.title("Numero de video juegos realizados por año")
plt.xlabel("Año")
plt.ylabel("Juegos realizados")
plt.show()
```

```python
sns.boxplot(x=filter_year["Year_of_Release"])
```

##### 6.2 Graficando segunda variable genero.

Ya una vez filtrado el dataset con los años que estamos considerando, graficamos para ver la frecuencia de los generos que estamos interesados (Peleas y Action). 


```python

Genre_counts = filter_year.groupby("Genre").size()
plt.figure(figsize=(15,8))
plt.bar(Genre_counts.index, Genre_counts.values)
plt.title("Numero de juegos por Genero")
plt.xlabel("Genero del Juego")
plt.ylabel("Juegos realizados")
plt.show()
```

Podemos ver que la que predomina es Acción, y en pelea el número de videojuegos no es tan alto. 


##### 6.3 Verificando la cantidad de juegos que cumplen nuestras caracteristicas para cada consola. 

Ya habiendo visto que el dataset cumple con nuestras variables, podemos visualizar solo la información que nos es util, y graficar con respecto a las consolas.   

```python
filter_complate = filter_year[filter_year.Genre.isin(['Action','Fighting'])]  
```

```python
Genre_counts = filter_complate.groupby("Platform").size()
plt.figure(figsize=(15,8))
plt.bar(Genre_counts.index, Genre_counts.values)
plt.title("Numero de juegos por Genero")
plt.xlabel("Genero del Juego")
plt.ylabel("Juegos realizados")
plt.show()
```

Podemos ver que PS4 y PSV y XOne, son las consolas que mayormente cumplen nuestras caracteristicas.
